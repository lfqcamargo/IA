```python
import pandas as pd
import numpy  as np

dados = '/content/drive/MyDrive/Colab Notebooks/Dados/cirrhosis.csv'

df_original = pd.read_csv(dados)
df_original
```

```python
df_original.info()
```

```python
df_sem_nulos = df_original.dropna()
df_sem_nulos
```

```python
df_tratado = df_sem_nulos.drop(columns=['Status', 'ID'])
df_tratado
```

```python
colunas_categoricas = ['Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']
df_tratado.reset_index(inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='if_binary')

df_categorico = df_tratado[colunas_categoricas]
```

```python
df_encoded = pd.DataFrame(encoder.fit_transform(df_categorico).toarray(),
                          columns=encoder.get_feature_names_out(colunas_categoricas))

df_encoded
```

```python
df_final = pd.concat([df_tratado.drop(columns=colunas_categoricas), df_encoded], axis=1)
df_final.drop(columns='index', inplace=True)
df_final
```

```python
X = df_final.drop(columns='Stage')
y = df_final['Stage']
```

```python
from sklearn.model_selection import train_test_split

X, X_teste, y, y_teste = train_test_split(X,y, random_state=42, stratify=y, test_size=0.15)
X_treino, X_val, y_treino, y_val = train_test_split(X,y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo = RandomForestClassifier(max_depth=7, random_state=42)
modelo.fit(X_treino,y_treino)
y_pred = modelo.predict(X_val)

print(f'Acurácia de Treino: {modelo.score(X_treino,y_treino) * 100}')
print(f'Acurácia de Validação: {modelo.score(X_val,y_val) * 100}')
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_val,y_pred)
print('Relátorio de Classificação:\n', report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_val, y_pred, normalize='true', cmap='Blues')
```

```python
from sklearn.dummy import DummyClassifier

modelo_dummy = DummyClassifier(strategy='uniform')
modelo_dummy.fit(X_treino,y_treino)
y_pred_dummy = modelo_dummy.predict(X_val)

print(f'Acurácia Dummy de Treino: {modelo_dummy.score(X_treino,y_treino) * 100}')
print(f'Acurácia Dummy de Validação: {modelo_dummy.score(X_val,y_val) * 100}')
```

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(random_state=42)
X_balanceado, y_balanceado = oversample.fit_resample(X_treino,y_treino)
y_balanceado.value_counts()
```

```python
modelo_rf = RandomForestClassifier(random_state=42, max_depth=5)
modelo_rf.fit(X_balanceado,y_balanceado)

y_pred = modelo_rf.predict(X_val)
print(f'Acurácia de treino: {modelo_rf.score(X_balanceado,y_balanceado)}')
print(f'Acurácia de validação: {modelo_rf.score(X_val,y_val)}')
```

```python
report = classification_report(y_val, y_pred)
print('Relátorio de Classificação:\n', report)
ConfusionMatrixDisplay.from_predictions(y_val, y_pred, normalize='true', cmap='Blues');
```

```python
from imblearn.pipeline import Pipeline as imbpipeline

modelo = RandomForestClassifier(max_depth=7)
pipeline = imbpipeline([('oversample', RandomOverSampler()), ('floresta', modelo)])
```

```python
from sklearn.model_selection import cross_validate, StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_resultados = cross_validate(pipeline, X, y, cv=skf, scoring='recall_weighted')
cv_resultados['test_score']
```
