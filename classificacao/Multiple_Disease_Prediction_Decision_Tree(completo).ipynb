# Descrição do Conjunto de Dados de Saúde

Este conjunto de dados é desenvolvido para verificar a saúde de uma pessoa, especificamente se ela possui uma doença específica ou está em bom estado de saúde, com base em amostras de sangue ou outros parâmetros. É importante notar que:

- Os dados são **criados manualmente** e destinam-se **exclusivamente para fins de aprendizado**.
- A **pesquisa foi realizada** sobre a porcentagem de cada parâmetro e os possíveis impactos (altos ou baixos) na saúde.
- **Doenças potenciais** associadas a níveis altos ou baixos desses parâmetros foram identificadas com base nesta pesquisa.
- Com base nessa informação e nas pesquisas realizadas, este conjunto de dados foi criado para fins educacionais e como uma sugestão de projeto.

## Características dos Dados

- **Escalonamento**: Os dados já estão escalonados no intervalo de `0` a `1`, facilitando a análise e modelagem.

## Objetivo

O principal objetivo deste conjunto de dados é fornecer um recurso para o desenvolvimento e teste de projetos de aprendizado de máquina, permitindo aos usuários explorar a relação entre diferentes parâmetros sanguíneos e o estado de saúde de uma pessoa.

## Uso

Este conjunto de dados é ideal para quem está procurando **praticar habilidades de análise de dados**, desenvolver modelos de **aprendizado de máquina** ou explorar **relações médicas** entre parâmetros sanguíneos e doenças específicas. Lembre-se de que os dados são puramente educacionais e não devem ser usados para diagnóstico médico real.


# Health Parameters Range

This dataset includes a range of health parameters which could indicate the health status of a person. Below are the normal ranges for each parameter:

- **Glucose**: (70, 140) mg/dL
- **Cholesterol**: (125, 200) mg/dL
- **Hemoglobin**: (13.5, 17.5) g/dL
- **Platelets**: (150,000, 450,000) per microliter of blood
- **White Blood Cells**: (4,000, 11,000) per cubic millimeter of blood
- **Red Blood Cells**: (4.2, 5.4) million cells per microliter of blood
- **Hematocrit**: (38, 52)%
- **Mean Corpuscular Volume**: (80, 100) femtoliters
- **Mean Corpuscular Hemoglobin**: (27, 33) picograms
- **Mean Corpuscular Hemoglobin Concentration**: (32, 36) grams per deciliter
- **Insulin**: (5, 25) microU/mL
- **BMI**: (18.5, 24.9) kg/m^2
- **Systolic Blood Pressure**: (90, 120) mmHg
- **Diastolic Blood Pressure**: (60, 80) mmHg
- **Triglycerides**: (50, 150) mg/dL
- **HbA1c**: (4, 6)%
- **LDL Cholesterol**: (70, 130) mg/dL
- **HDL Cholesterol**: (40, 60) mg/dL
- **ALT**: (10, 40) U/L
- **AST**: (10, 40) U/L
- **Heart Rate**: (60, 100) beats per minute
- **Creatinine**: (0.6, 1.2) mg/dL
- **Troponin**: (0, 0.04) ng/mL
- **C-reactive Protein**: (0, 3) mg/L

Please note that these ranges are considered normal for most adults. However, individual circumstances and medical conditions may alter what's considered normal for a specific individual. Always consult healthcare professionals for personal medical advice.

```python
import pandas as pd
import numpy  as np

dados = '/content/drive/MyDrive/Colab Notebooks/Dados/Blood_samples_dataset_balanced_2(f).csv'

df_original = pd.read_csv(dados)
df_original
```

```python
from sklearn.preprocessing import LabelEncoder

df_one_hot = pd.get_dummies(df_original, columns=['Disease'])

label_encoder = LabelEncoder()
df_encoded = df_original.copy()
df_encoded['Disease_encoded'] = label_encoder.fit_transform(df_encoded['Disease'])
df_encoded
```

```python
def verificar_categorias():
  diseases = ['Healthy', 'Diabetes', 'Thalasse', 'Anemia', 'Thromboc']

  label_encoder = LabelEncoder()
  disease_encoded = label_encoder.fit_transform(diseases)

  for i, category in enumerate(label_encoder.classes_):
      print(f"Categoria '{category}' foi codificada como {i}.")
```

```python
verificar_categorias()
```

```python
df_encoded.drop(columns='Disease', inplace=True)
df_encoded
```

```python
X = df_encoded.drop(columns='Disease_encoded')
y = df_encoded['Disease_encoded']
```

```python
from sklearn.model_selection import train_test_split

RANDOM_SEED = 42

X, X_teste, y, y_teste = train_test_split(X, y, test_size=0.20, stratify=y, random_state=RANDOM_SEED)
X_treino, X_val, y_treino, y_val = train_test_split(X, y, stratify=y, random_state=RANDOM_SEED)
```

```python
#Definindo Modelo Base Most_Frequent
from sklearn.dummy import DummyClassifier

modelo_most_frequent = DummyClassifier(strategy='most_frequent')
modelo_most_frequent.fit(X_treino, y_treino)

y_pred_most_frequent = modelo_most_frequent.predict(X_val)
y_val_most_frequent  = modelo_most_frequent.predict(X_teste)
print(f'Acurácia de Treino:    {modelo_most_frequent.score(X_treino, y_treino) * 100}')
print(f'Acurácia de Validação: {modelo_most_frequent.score(X_val, y_val) * 100}')
print(f'Acurácia de Teste:     {modelo_most_frequent.score(X_teste, y_teste) * 100}')
```

```python
#Definindo Modelo Base Stratified
from sklearn.dummy import DummyClassifier

modelo_stratified = DummyClassifier(strategy='stratified')
modelo_stratified.fit(X_treino, y_treino)

y_pred_stratified = modelo_stratified.predict(X_val)
y_val_stratified  = modelo_stratified.predict(X_teste)
print(f'Acurácia de Treino:    {modelo_stratified.score(X_treino, y_treino) * 100}')
print(f'Acurácia de Validação: {modelo_stratified.score(X_val, y_val) * 100}')
print(f'Acurácia de Teste:     {modelo_stratified.score(X_teste, y_teste) * 100}')
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo_dt = DecisionTreeClassifier(max_depth=3, random_state=RANDOM_SEED)
modelo_dt.fit(X_treino, y_treino)
```

```python
y_pred_modelo_dt = modelo_dt.predict(X_val)
print(f'Acurácia de Treino Modelo Decison Tree: {modelo_dt.score(X_treino, y_treino) * 100}')
print(f'Acurácia de Validação Modelo Decision Tree: {modelo_dt.score(X_val, y_val) * 100}')
```

```python
display_labels = ['Anemia', 'Diabetes', 'Healthy', 'Thalasse', 'Thromboc']
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report

report_modelo_dt = classification_report(y_val, y_pred_modelo_dt)
print('Relátorio Classificação Modelo Decision Tree\n', report_modelo_dt);

matriz_confusao_dt = confusion_matrix(y_val, y_pred_modelo_dt)
visualizacao_dt = ConfusionMatrixDisplay(confusion_matrix=matriz_confusao_dt, display_labels=display_labels)
ConfusionMatrixDisplay.from_predictions(y_val, y_pred_modelo_dt, normalize='true', cmap='Blues', display_labels=display_labels);
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

inner_cv = StratifiedKFold(shuffle=True, random_state=RANDOM_SEED)
outer_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_SEED)

param_grid_dt = {
    'criterion':  ['gini', 'entropy'],
    'max_depth': np.linspace(6, 12, 4, dtype=int),
    'min_samples_split': np.linspace(5, 20, 4, dtype=int),
    'min_samples_leaf': np.linspace(5, 20, 4, dtype=int),
    'max_features': ['sqrt', 'log2'],
    'splitter': ['best', 'random']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)

dt_grid_search = GridSearchCV(estimator = DecisionTreeClassifier(random_state=RANDOM_SEED),
                              param_grid = param_grid_dt,
                              scoring = "accuracy",
                              n_jobs = -1,
                              cv = inner_cv)

dt_nested_scores = cross_val_score(dt_grid_search, X_treino, y_treino, cv=outer_cv)
```

```python
print(f'Resultado de Cada Iteração do CV Externo: {dt_nested_scores}')
print(f'Média: {dt_nested_scores.mean()}')
```

```python
dt_grid_search.fit(X_treino, y_treino)
melhores_parametros_dt = dt_grid_search.best_params_
melhor_score_dt        = dt_grid_search.best_score_
print("Melhores parâmetros:", melhores_parametros_dt)
print("Melhor score:", melhor_score_dt)
```

```python
modelo_dt_final = DecisionTreeClassifier(criterion='gini',
                                         max_depth=10,
                                         max_features='sqrt',
                                         min_samples_leaf=5,
                                         min_samples_split=5,
                                         splitter='best')
modelo_dt_final.fit(X_treino, y_treino)
y_pred_modelo_dt_final = modelo_dt_final.predict(X_val)
print(f'Acurácia de treino: {modelo_dt_final.score(X_treino,y_treino)}')
print(f'Acurácia de validação: {modelo_dt_final.score(X_val,y_val)}')
```

```python
y_teste_modelo_dt_final = modelo_dt_final.predict(X_teste)
print(f'Acurácia de Teste: {modelo_dt_final.score(X_teste,y_teste)}')
```

```python
report_modelo_dt_final = classification_report(y_teste, y_teste_modelo_dt_final)
print('Relátorio Classificação Modelo Decision Tree Final\n', report_modelo_dt_final);

matriz_confusao_dt_final = confusion_matrix(y_teste, y_teste_modelo_dt_final)
visualizacao_dt_final = ConfusionMatrixDisplay(confusion_matrix=matriz_confusao_dt, display_labels=display_labels)
ConfusionMatrixDisplay.from_predictions(y_teste, y_teste_modelo_dt_final, normalize='true', cmap='Blues', display_labels=display_labels);
```
